```python
from sympy import symbols, Eq, solve
```

```python
# Definir variables simbólicas
x, y, z = symbols('x y z')

print("Resolviendo un sistema 3x3 con el MÉTODO DE SUSTITUCIÓN")
print("Forma general:")
print("a1*x + b1*y + c1*z = d1")
print("a2*x + b2*y + c2*z = d2")
print("a3*x + b3*y + c3*z = d3")
print("-" * 50)

# Pedimos los coeficientes al usuario
a1, b1, c1, d1 = map(float, input("Ingrese los coeficientes de la primera ecuación (a1 b1 c1 d1): ").split())
a2, b2, c2, d2 = map(float, input("Ingrese los coeficientes de la segunda ecuación (a2 b2 c2 d2): ").split())
a3, b3, c3, d3 = map(float, input("Ingrese los coeficientes de la tercera ecuación (a3 b3 c3 d3): ").split())

# Crear ecuaciones simbólicas
ec1 = Eq(a1*x + b1*y + c1*z, d1)
ec2 = Eq(a2*x + b2*y + c2*z, d2)
ec3 = Eq(a3*x + b3*y + c3*z, d3)

print("\nEcuaciones ingresadas:")
print("E1:", ec1)
print("E2:", ec2)
print("E3:", ec3)

# Paso 1: Despejar x de la primera ecuación
despeje_x1_list = solve(ec1, x)
if not despeje_x1_list:
    print("No se pudo despejar x de la primera ecuación.")
    exit()
despeje_x1 = despeje_x1_list[0]
print(f"\nPaso 1: Despejamos x de la primera ecuación:")
print(f"x = {despeje_x1}")

# Paso 2: Despejar x de la segunda ecuación
despeje_x2_list = solve(ec2, x)
if not despeje_x2_list:
    print("No se pudo despejar x de la segunda ecuación.")
    exit()
despeje_x2 = despeje_x2_list[0]
print(f"\nPaso 2: Despejamos x de la segunda ecuación:")
print(f"x = {despeje_x2}")


# Paso 3: Igualar las dos expresiones de x y resolver y
equal1 = Eq(despeje_x1, despeje_x2)
sol_y1_list = solve(equal1, y)
if not sol_y1_list:
    print("No se pudo resolver y igualando x1 y x2.")
    exit()
sol_y1 = sol_y1_list[0]
print(f"\nPaso 3: Igualamos x1 y x2 para resolver y:")
print(f"y = {sol_y1}")


# Paso 4: Despejar x de la tercera ecuación
despeje_x3_list = solve(ec3, x)
if not despeje_x3_list:
    print("No se pudo despejar x de la tercera ecuación.")
    exit()
despeje_x3 = despeje_x3_list[0]
print(f"\nPaso 4: Despejamos x de la tercera ecuación:")
print(f"x = {despeje_x3}")


# Paso 5: Igualar x2 y x3 y resolver y
equal2 = Eq(despeje_x2, despeje_x3)
sol_y2_list = solve(equal2, y)
if not sol_y2_list:
    print("No se pudo resolver y igualando x2 y x3.")
    exit()
sol_y2 = sol_y2_list[0]
print(f"\nPaso 5: Igualamos x2 y x3 para resolver y:")
print(f"y = {sol_y2}")


# Paso 6: Igualar las dos expresiones de y y resolver z
equaly = Eq(sol_y1, sol_y2)
sol_z_list = solve(equaly, z)
if not sol_z_list:
    print("No se pudo resolver z igualando y1 y y2.")
    exit()
sol_z = sol_z_list[0]
print(f"\nPaso 6: Igualamos y1 y y2 para resolver z:")
print(f"z = {sol_z}")


# Paso 7: Sustituimos z en una de las ecuaciones de y
sol_y_list = solve(equal2.subs(z, sol_z), y)
if not sol_y_list:
    print("No se pudo resolver y sustituyendo z.")
    exit()
sol_y = sol_y_list[0]
print(f"\nPaso 7: Sustituimos z en una ecuación de y para obtener y:")
print(f"y = {sol_y}")

# Paso 8: Sustituimos y y z en una ecuación de x
sol_x_list = solve(ec1.subs([(y, sol_y), (z, sol_z)]), x)
if not sol_x_list:
    print("No se pudo resolver x sustituyendo y y z.")
    exit()
sol_x = sol_x_list[0]
print(f"\nPaso 8: Sustituimos y y z en una ecuación de x para obtener x:")
print(f"x = {sol_x}")

# Solución final
print("\n✅ Solución del sistema:")
print(f"x = {sol_x}, y = {sol_y}, z = {sol_z}")
```
